```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/advertising (1).csv')
```

```python
df.head(10)
```

```python
print(df.describe())
```

```python
fig,axs=plt.subplots(3,figsize=(5,5))
plt1=sns.distplot(df['TV'],ax=axs[0])
plt2=sns.boxplot(df['Newspaper'],ax=axs[2])
plt3=sns.boxplot(df['Radio'],ax=axs[1])
plt.tight_layout()
```

```python
X=df['TV']
Y=df['Sales']
```

```python
plt.scatter(X,Y)
plt.xlabel('TV')
plt.ylabel('Sales')
```

```python
 plt.scatter(X,Y,color='blue',label='scatter plot')
 plt.title('relationship between TV and Sales')
 plt.xlabel('TV')
 plt.ylabel('Sales')
 plt.legend()
 plt.show()
```

```python
X.shape
Y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=30,random_state=42)
```

```python

print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python
lm=lm.fit(X_train.values.reshape(-1, 1),Y_train)
Y_pred=lm.predict(X_test.values.reshape(-1, 1))
```

```python
plt.scatter(X_train,Y_train,color='lightblue')
plt.plot(X_train,lm.predict(X_train.values.reshape(-1, 1)),color='blue')
plt.title('Relationship between TV and Sales')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```
